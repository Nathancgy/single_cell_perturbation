```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset

def mae(x):
    return np.abs(x)

def mse(x):
    return x**2

def log_cosh(x):
    return np.log(np.cosh(x))

# Create a range of error values
x = np.linspace(-15, 15, 800)

# Calculate loss values for each function
mae_loss = mae(x)
mse_loss = mse(x)
log_cosh_loss = log_cosh(x)

# Create the plot
fig, ax = plt.subplots(figsize=(12, 8))
plt.style.use('default')

# Define new color scheme with purple, orange, and cyan
purple = '#8A2BE2'
orange = '#FFA500'
cyan = '#00FFFF'

# Plot the loss functions
ax.plot(x, mae_loss, label='MAE', linewidth=2.5, color=purple)
ax.plot(x, mse_loss, label='MSE', linewidth=2.5, color=orange)
ax.plot(x, log_cosh_loss, label='Log Cosh', linewidth=2.5, color=cyan)

# Fill areas under the curves
ax.fill_between(x, mae_loss, alpha=0.2, color=purple)
ax.fill_between(x, mse_loss, alpha=0.2, color=orange)
ax.fill_between(x, log_cosh_loss, alpha=0.2, color=cyan)

# Customize the plot
ax.set_title('MAE, MSE, and Log Cosh Loss Functions', fontsize=16, fontweight='bold')
ax.set_xlabel('Error', fontsize=19)
ax.set_ylabel('Loss', fontsize=19)
ax.legend(fontsize=15)
ax.grid(True, linestyle='--', alpha=0.7)

# Set y-axis limit to better show the differences
ax.set_ylim(0, 50)

# Add annotations
ax.annotate('MAE: Linear growth', xy=(-8, 8), xytext=(-9, 15),
            arrowprops=dict(facecolor=purple, shrink=0.05), fontsize=15, color=purple)
ax.annotate('MSE: Quadratic growth', xy=(8, 40), xytext=(5, 45),
            arrowprops=dict(facecolor=orange, shrink=0.05), fontsize=15, color=orange)
ax.annotate('Log Cosh: Smooth transition', xy=(0, 0.5), xytext=(2, 5),
            arrowprops=dict(facecolor=cyan, shrink=0.05), fontsize=15, color=cyan)

# Create a zoomed inset
axins = zoomed_inset_axes(ax, zoom=6, loc='center right')
axins.plot(x, mae_loss, color=purple, linewidth=2.5)
axins.plot(x, mse_loss, color=orange, linewidth=2.5)
axins.plot(x, log_cosh_loss, color=cyan, linewidth=2.5)

# Set the range for the zoomed region
x1, x2, y1, y2 = -2, 2, 0, 2
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
axins.grid(True, linestyle='--', alpha=0.7)

# Remove ticks for the inset plot
axins.tick_params(labelleft=False, labelbottom=False)

# Draw a box around the zoomed region
mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="0.5")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def bce_loss(y_true, y_pred):
    epsilon = 1e-15  # Small value to avoid log(0)
    return -(y_true * np.log(y_pred + epsilon) + (1 - y_true) * np.log(1 - y_pred + epsilon))

# Generate data
y_true = np.linspace(0, 1, 1000)
y_pred_ranges = [0.1, 0.3, 0.5, 0.7, 0.9]

# Create the plot
plt.figure(figsize=(12, 6))
plt.style.use('default')  # Use default style instead of seaborn-whitegrid

# Plot BCE loss for different predicted probabilities
for y_pred in y_pred_ranges:
    loss = bce_loss(y_true, y_pred)
    plt.plot(y_true, loss, label=f'y_pred = {y_pred}', linewidth=2)

# Customize the plot
plt.title('Binary Cross-Entropy Loss', fontsize=20, fontweight='bold')
plt.xlabel('True Value', fontsize=18)
plt.ylabel('Loss', fontsize=18)
plt.legend(fontsize=14, title='Predicted Probability')
plt.grid(True, linestyle='--', alpha=0.7)

# Set axis limits
plt.xlim(0, 1)
plt.ylim(0, 5)

# Add annotations
plt.annotate('High loss for\nincorrect predictions', xy=(0.1, 2.5), xytext=(0.3, 3.5),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=16)
plt.annotate('Low loss for\ncorrect predictions', xy=(0.9, 0.5), xytext=(0.6, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=16)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from tqdm import tqdm

# Load the data
print("Loading data...")
visualization_data = pd.read_parquet('qsar_modeling/visualization_data.parquet')

# Separate fingerprints and QSAR features
fingerprint_cols = [col for col in visualization_data.columns if col.startswith('fp_')]
qsar_cols = [col for col in visualization_data.columns if col.startswith('feature_')]

X = visualization_data[fingerprint_cols].values
qsar_features = visualization_data[qsar_cols]

# Perform t-SNE
print("Performing t-SNE...")
tsne = TSNE(n_components=2, random_state=42, n_jobs=-1)
X_tsne = tsne.fit_transform(X)

# Function to create and save t-SNE plot
def plot_tsne(feature_name):
    plt.figure(figsize=(12, 10))
    scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=qsar_features[feature_name], cmap='viridis', alpha=0.6)
    plt.colorbar(scatter, label=feature_name)
    plt.title(f't-SNE Plot of Compounds Colored by {feature_name}')
    plt.xlabel('t-SNE 1')
    plt.ylabel('t-SNE 2')
    plt.tight_layout()
    plt.savefig(f'tsne_plot_{feature_name}.png', dpi=300)
    plt.close()

# Create t-SNE plots for the first 5 QSAR features
print("Creating t-SNE plots...")
for feature in tqdm(qsar_cols[:5], desc="Plotting"):
    plot_tsne(feature)

print("t-SNE plots created and saved.")

# Optional: Create a combined plot of the first 4 QSAR features
print("Creating combined t-SNE plot...")
fig, axs = plt.subplots(2, 2, figsize=(20, 20))
for i, feature in enumerate(qsar_cols[:4]):
    row = i // 2
    col = i % 2
    scatter = axs[row, col].scatter(X_tsne[:, 0], X_tsne[:, 1], c=qsar_features[feature], cmap='viridis', alpha=0.6)
    axs[row, col].set_title(f'{feature}')
    axs[row, col].set_xlabel('t-SNE 1')
    axs[row, col].set_ylabel('t-SNE 2')
    fig.colorbar(scatter, ax=axs[row, col])

plt.tight_layout()
plt.savefig('tsne_plot_combined.png', dpi=300)
plt.close()

print("Combined t-SNE plot created and saved.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from tqdm import tqdm

# Load the data
print("Loading data...")
visualization_data = pd.read_parquet('qsar_modeling/visualization_data.parquet')
feature_importance = np.load('qsar_modeling/qsar_models/feature_importance.npy')
test_predictions = np.load('qsar_modeling/qsar_models/test_predictions.npy')
test_actual = np.load('qsar_modeling/qsar_models/test_actual.npy')

# Identify top QSAR properties based on variance
qsar_cols = [col for col in visualization_data.columns if col.startswith('feature_')]
top_features = visualization_data[qsar_cols].var().nlargest(15).index.tolist()

# 1. Feature Importance Heatmap
print("Generating Feature Importance Heatmap...")
plt.figure(figsize=(12, 8))
top_feature_indices = [int(feature.split('_')[1]) for feature in top_features]
sns.heatmap(feature_importance[top_feature_indices, :20], cmap='viridis', annot=False)
plt.title('Feature Importance Heatmap (Top 15 QSAR Properties x Top 20 Fingerprint Bits)')
plt.xlabel('Morgan Fingerprint Bit', fontsize=14)
plt.ylabel('QSAR Property', fontsize=14)
plt.yticks(range(15), top_features, rotation=0)
plt.show()

# 2. PCA vs t-SNE Comparison
print("Generating PCA vs t-SNE Comparison...")
fingerprint_cols = [col for col in visualization_data.columns if col.startswith('fp_')]
X = visualization_data[fingerprint_cols].values

pca = PCA(n_components=2)
tsne = TSNE(n_components=2, random_state=42)

X_pca = pca.fit_transform(X)
X_tsne = tsne.fit_transform(X)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
scatter1 = ax1.scatter(X_pca[:, 0], X_pca[:, 1], c=visualization_data[top_features[0]], cmap='viridis', alpha=0.6)
ax1.set_title('PCA', fontsize=24)
ax1.set_xlabel('PC1', fontsize=24)
ax1.set_ylabel('PC2', fontsize=24)
fig.colorbar(scatter1, ax=ax1, label=top_features[0])

scatter2 = ax2.scatter(X_tsne[:, 0], X_tsne[:, 1], c=visualization_data[top_features[0]], cmap='viridis', alpha=0.6)
ax2.set_title('t-SNE', fontsize=24)
ax2.set_xlabel('t-SNE 1', fontsize=24)
ax2.set_ylabel('t-SNE 2', fontsize=24)
fig.colorbar(scatter2, ax=ax2, label=top_features[0])

plt.tight_layout()
plt.show()

# 3. Correlation Matrix of Top QSAR Properties
print("Generating Correlation Matrix of Top QSAR Properties...")
correlation_matrix = visualization_data[top_features].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Matrix of Top 15 QSAR Properties')
plt.tight_layout()
plt.show()

# 4. Actual vs Predicted Scatter Plot
print("Generating Actual vs Predicted Scatter Plot...")
fig, axes = plt.subplots(2, 2, figsize=(20, 20))

def remove_outliers(x, y):
    combined = np.column_stack((x, y))
    q1 = np.percentile(combined, 25, axis=0)
    q3 = np.percentile(combined, 75, axis=0)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    mask = np.all((combined > lower_bound) & (combined < upper_bound), axis=1)
    return x[mask], y[mask]

for i, feature in enumerate(top_features[:4]):
    ax = axes[i//2, i%2]
    feature_index = int(feature.split('_')[1])
    
    # Remove outliers
    actual, predicted = remove_outliers(test_actual[:, feature_index], test_predictions[:, feature_index])
    
    scatter = ax.scatter(actual, predicted, alpha=0.5)
    
    # Calculate and plot the diagonal line
    min_val = min(actual.min(), predicted.min())
    max_val = max(actual.max(), predicted.max())
    ax.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2)
    
    ax.set_title(feature, fontsize=28)
    ax.set_xlabel('Actual', fontsize=28)
    ax.set_ylabel('Predicted', fontsize=28)
    
    # Set equal aspect ratio
    ax.set_aspect('equal', adjustable='box')
    
    # Add R-squared value
    r_squared = np.corrcoef(actual, predicted)[0, 1]**2
    ax.text(0.05, 0.95, f'R² = {r_squared:.3f}', transform=ax.transAxes, 
            verticalalignment='top', fontsize=12)

plt.tight_layout()
plt.show()

# 5. Distribution of QSAR Property Values
print("Generating Distribution of QSAR Property Values...")
fig, axes = plt.subplots(2, 2, figsize=(20, 20))
for i, feature in enumerate(top_features[:4]):
    ax = axes[i//2, i%2]
    sns.histplot(visualization_data[feature], kde=True, ax=ax)
    ax.set_title(f'Distribution of {feature}')
    ax.set_xlabel('Value', fontsize=14)
    ax.set_ylabel('Frequency', fontsize=14)
plt.tight_layout()
plt.show()

print("Visualizations completed.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score
from tqdm import tqdm

# Load the data
test_predictions = np.load('qsar_modeling/qsar_models/test_predictions.npy')
test_actual = np.load('qsar_modeling/qsar_models/test_actual.npy')

# Calculate R-squared values for all properties
r2_scores = []
invalid_properties = []

for i in tqdm(range(362), desc="Calculating R-squared"):
    r2 = r2_score(test_actual[:, i], test_predictions[:, i])
    if 0 <= r2 <= 1:
        r2_scores.append(r2)
    else:
        invalid_properties.append((i, r2))
        r2_scores.append(np.nan)  # Use NaN for invalid values

# Print information about invalid properties
if invalid_properties:
    print("The following properties have invalid R-squared values:")
    for prop, r2 in invalid_properties:
        print(f"Property_{prop}: R-squared = {r2}")
    print("These properties will be excluded from the plot.")

# Create a DataFrame for easier plotting
df_r2 = pd.DataFrame({'Property': [f'Property_{i}' for i in range(362)], 'R-squared': r2_scores})

# Remove invalid R-squared values
df_r2_valid = df_r2.dropna().sort_values('R-squared', ascending=False).reset_index(drop=True)

# Set up the plot style
plt.style.use('default')
sns.set_palette("viridis")

# Create the figure and axis
fig, ax = plt.subplots(figsize=(20, 10))

# Create the bar plot
bars = sns.barplot(x='Property', y='R-squared', data=df_r2_valid, ax=ax)

# Customize the plot
ax.set_title('R-squared Values for QSAR Properties', fontsize=35, pad=20)
ax.set_xlabel('QSAR Properties (sorted by R-squared)', fontsize=32, labelpad=15)
ax.set_ylabel('R-squared', fontsize=32, labelpad=15)

# Remove x-axis labels as there are too many
ax.set_xticks([])
ax.set_xticklabels([])

# Add a color gradient to the bars
for i, bar in enumerate(bars.patches):
    bar.set_facecolor(plt.cm.viridis(i / len(bars.patches)))

# Add text for highest and lowest R-squared values
ax.text(0, df_r2_valid['R-squared'].max(), f"Highest: {df_r2_valid['R-squared'].max():.3f}", 
        verticalalignment='bottom', horizontalalignment='left', fontsize=22, fontweight='bold')
ax.text(len(df_r2_valid) - 1, df_r2_valid['R-squared'].min(), f"Lowest: {df_r2_valid['R-squared'].min():.3f}", 
        verticalalignment='top', horizontalalignment='right', fontsize=22, fontweight='bold')

# Add mean R-squared line
mean_r2 = df_r2_valid['R-squared'].mean()
ax.axhline(mean_r2, color='red', linestyle='--', linewidth=2)
ax.text(len(df_r2_valid) / 2, mean_r2, f"Mean: {mean_r2:.3f}", 
        verticalalignment='bottom', horizontalalignment='center', fontsize=22, fontweight='bold', color='red')

# Improve the layout
plt.tight_layout()

# Show the plot
plt.show()

print(f"R-squared visualization completed. {len(df_r2_valid)} valid properties plotted.")
```
